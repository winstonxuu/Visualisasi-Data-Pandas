# Week 13 - Membuat kolom baru dengan looping dan f-string

```python
import pandas as pd

print(pd.__version__)
```

```python
d = {'penjual':['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'], 
     'barang':['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']}

df = pd.DataFrame(d)
df
```

```python
cols = ['penjual', 'barang']

for col in cols:
    df[f'count_tiap_{col}'] =  df.groupby(col).cumcount() + 1
    
df
```

# Week 13 - Seleksi baris dengan between ( )

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), 
                  columns=cols)
df
```

```python
df[df['A'].between(2, 5)]
```

```python
df[df['A'].between(2, 5, inclusive=False)]
```

# Week 13 - Transformasi kolom menjadi baris pada Data Frame


```python
import pandas as pd

print(pd.__version__)
```

```python
d = {
'kode_area': [123, 456, 789, 321],
'pabrik': [4, 2, 5, 0],
'gudang': [2, 4, 7, 3],
'toko': [64, 32, 15, 24]
}

df = pd.DataFrame(d)
df
```

```python
df = df.melt(id_vars='kode_area', 
             var_name='jenis_bangunan', 
             value_name='jumlah')
df
```

# Week 11 - Menerapkan Agregasi Pada Sejumlah Kolom Dengan agg()

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
df.groupby('pclass').agg({'pclass':'count', 
                          'age':['mean', 'max'], 
                          'survived': 'mean'})
```

```python
df.groupby('pclass').agg(n_pass=('pclass', 'count'),
                         avg_age=('age', 'mean'),
                         max_age=('age', 'max'), 
                         survival_rate=('survived', 'mean'))
```

# Week 11 - Menangani whitespace pada Data Frame

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
data = {'nim': ['10', '11', '12', '13', '  '],
        'nama': ['adi', '  ', 'tejo', '  ', 'bejo']}

df = pd.DataFrame(data)
df
```

```python
df
```

```python
df.info()
```

```python
df = df.replace(r'^\s*$', np.nan, regex=True)
df
```

```python
df.info()
```

# Week 11 - Menata ulang penempatan kolom pada Data Frame

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

```python
df[['D', 'C', 'A', 'E', 'B']] #penataan ulang ini sifatnya tidak inplace, artinya data frame asli di variable df belum berubah
```

```python
df
```

```python
df = df[['D', 'C', 'A', 'E', 'B']] #ini cara merubah data frame di variable df, yakni dengan memasukan urutan df yang baru ke df yang lama
df
```

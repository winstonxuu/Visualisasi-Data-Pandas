# Week 10 - Seleksi baris dengan banyak kriteria

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
df[(df['sex']=='female') & (df['age']>=60) & (df['embarked']=='S') & (df['survived']==1)]
```

```python
df[
    (df['sex']=='female') & 
    (df['age']>=60) & 
    (df['embarked']=='S') & 
    (df['survived']==1)
]
```

```python
kr1 = df['sex']=='female'
kr2 = df['age']>=60
kr3 = df['embarked']=='S'
kr4 = df['survived']==1

df[kr1 & kr2 & kr3 & kr4]
```

# Week 10 - Mengakses sekelompok data dengan get_group()

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.read_csv('titanicfull.csv')
df.head()
```

```python
grouped_df = df.groupby('sex')
```

```python
grouped_df.get_group('female').head(10)
```

```python
grouped_df = df.groupby('survived')
```

```python
grouped_df.get_group(1).head(10)
```
